 **Objective 1 : Art and Design apps have better rating than others**

````python
# Grouping Category and Rating columns with count function, also dropping ['1.9' in Category,'19.0' in Rating] values in this which is invalid.
playstore_app = dfall.groupby(['Category','Rating'])['Rating'].count().unstack().drop([19.0], axis = 1).drop(['1.9'], axis = 0)

# Separating the Rating above 4.0 which will be considered as the best
separated_data = playstore_app.iloc[:,-10:]

# Taking Number of Rating above 4.0 using Sum method
df1 = separated_data.sum(axis = 1)
# Changing the above column into a Series
series1 = pd.Series(df1)

# Taking the Total number of Rating for each Category
total_rating = playstore_app.sum(axis = 1)
# Changing the above column into a Series
series2 = pd.Series(total_rating)

# Making an Average for each Category to see which Tops the board
Average_rating = ((df1 / total_rating)*100).round()
# Changing the above column into a Series
series3 = pd.Series(Average_rating)

# Concating Three datas to get a better look 
Apps_total_rating = pd.concat([series1,series2,series3], axis = 1)
# Renaming the columns
Apps_total_rating.rename(columns = {0:'Ratings > 4.0',1:'Total Ratings',2:'Average Rating'}, inplace = True)
````

**Objective 8: Number of installs as per the App size**

````python
# Separating the Category - Game, from that taking columns Size and Installs.
''' Replacing the cells that have 
                    (a) 'Varies with device' with '0' in Size column
                    (b) Special characters 
                    (c) Size column also contains M denoting MB of the app,so making that as a space
                    (d) Size column also contains k denoting KB of the app,so making that as a '0' to differentiate from other apps
                    and resetting the index.'''

df8 = (dfall.loc[dfall['Category'] == 'GAME'].loc[:,['Size','Installs']].replace('Varies with device','0').replace('\W','',regex = True)
                     .replace('M','',regex = True).replace('k','0',regex = True).reset_index(drop = True))

# Changing the type of columns Installs and Size 
df8.Size = df8.Size.astype(int)
df8.Installs = df8.Installs.astype(int)

# Segregating the Apps according to the Size and also using Sum method to know the sum of number of Installs
df8_1 = df8.loc[(df8['Size'] >= 1 ) & (df8['Size']<=25)].sum()

df8_2= df8.loc[(df8['Size']>25) & (df8['Size'] < 50)].sum()

df8_3= df8.loc[(df8['Size']>=50) & (df8['Size'] < 100)].sum()

df8_4 = df8.loc[df8['Size']>=100].sum()

# The 'Varies with device' was replaced with 0, so taking that apps into consideration
df8_5 = df8.loc[df8['Size'] == 0].sum()

# Apps which are in KB size has been added with 0 at the end,so those values are above 1000.Taking that into consideration
df8_6 = df8.loc[df8['Size']>=1000].sum()

# Assigning Labels for each row
Categories = ['Below 25MB','25MB to 50MB','50MB to 100MB','Above 100MB','Varies with device','Apps in KB']

# Making the values as a Dataframe to align it 
df8_all = pd.DataFrame([df8_1,df8_2,df8_3,df8_4,df8_5,df8_6],index = Categories)

# We need Installs column alone,so separating that with labels 
df8_all_ = df8_all.loc[:,['Installs']]
````
